```python
import math

def caluculateprob(N, n):
  pattern = math.factorial(N) / (math.factorial(n) * math.factorial(N - n))
  return pattern

proborigin = (1 / caluculateprob(64, 32)) * 10**(19)

patternA = caluculateprob(32, 1)
patternB = caluculateprob(32, 2)
patternC = caluculateprob(32, 8)
patternD = caluculateprob(32, 14)
patternE = caluculateprob(32, 16)


wantA = proborigin * patternA * 10**(-2)
wantB = proborigin * patternB * 10**(-3)
wantC = proborigin * patternC * 10**(-7)
wantD = proborigin * patternD * 10**(-9)
wantE = proborigin * patternE * 10**(-9)

AE = wantA / wantE
BE = wantB / wantE
CE = wantC / wantE
DE = wantD / wantE

AE, BE, CE, DE

```
